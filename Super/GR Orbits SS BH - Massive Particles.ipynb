```python
import math
import numpy as np
import matplotlib.pyplot as plt
#initial conditions
u0 = 1/50
u1 = 1/50
#step size
dp = math.pi/200
#mass
m = 1
#number of steps
n = 1200
#
#J = 6 #ang mom
l = 1 #lambda
```

```python
def gr(x1, x0, J):
    x = (3*x1-1)*dp**2*x1+2*x1-x0+l*m**2/J**2*dp**2
    
    return x

def newton(x1, x0,J):
    x = (-1)*dp**2*x1+2*x1-x0+l*m**2/J**2*dp**2
    
    return x
```

```python
z = np.zeros(n);
p = np.zeros(n);
z[0] = u0
z[1] = u1
p[0] = 0
p[1] = dp
J = 5.5
#
z1 = np.zeros(n);
p1 = np.zeros(n);
z1[0] = 1/4
z1[1] = 1/4
p1[0] = 0
p1[1] = dp
J1 = 4
#
z2 = np.zeros(n);
p2 = np.zeros(n);
z2[0] = 1/6
z2[1] = 1/6
p2[0] = 0
p2[1] = dp
J2 = 4.2
```

```python
for i in range(2,n):
     z[i] = gr(z[i-1],z[i-2],J);
     p[i] = dp*i
```

```python
for i in range(2,n):
     z1[i] = gr(z1[i-1],z1[i-2],J1);
     p1[i] = dp*i
    
for i in range(2,n):
     z2[i] = gr(z2[i-1],z2[i-2],J2);
     p2[i] = dp*i
```

```python
plt.plot(p, m/z)
plt.plot(p1, m/z1, label = "1")
plt.plot(p1, m/z2, label="2")
plt.ylabel('r')
plt.xlabel('phi')
plt.legend()
plt.show()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(8, 8), dpi=80)
plt.plot(m*np.cos(p)/z,m*np.sin(p)/z, label="GR prediction")
plt.plot(m*np.cos(pn)/zn,m*np.sin(pn)/zn, label="Newtonian orbit")
#plt.plot(m*np.cos(p1)/z1,m*np.sin(p1)/z1)
#plt.plot(m*np.cos(p2)/z2,m*np.sin(p2)/z2)
plt.legend()
plt.ylabel('y')
plt.xlabel('x, [x] = m')
fig=plt.figure(1)
plt.axis([-50,50,-50,50])
ax=fig.add_subplot(1,1,1)
circ=plt.Circle((0.5,0.5), radius=0.001, color='black', fill=False,transform=ax.transAxes)
ax.add_patch(circ)
plt.grid()
plt.show()
```
