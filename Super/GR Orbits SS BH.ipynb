```python
import math
import numpy as np
#initial conditions
u0 = 1/2.5
u1 = 1/2.5
#step size
dp = math.pi/400
#mass
m = 1
#number of steps
n = 500
```

```python
def op(x1, x0):
    x = (3*x1-1)*dp**2*x1+2*x1-x0
    
    return x

def newton(x1, x0):
    x = (-1)*dp**2*x1+2*x1-x0
    
    return x
```

```python
z = np.zeros(n);
p = np.zeros(n);
z[0] = u0
z[1] = u1
p[0] = 0
p[1] = dp
#
z1 = np.zeros(2*n);
p1 = np.zeros(2*n);
z1[0] = 1/3
z1[1] = 1/3
p1[0] = 0
p1[1] = dp
#
z2 = np.zeros(n);
p2 = np.zeros(n);
z2[0] = 1/4
z2[1] = 1/4
p2[0] = 0
p2[1] = dp
```

```python
for i in range(2,n):
     z[i] = op(z[i-1],z[i-2]);
     p[i] = dp*i
```

```python
for i in range(2,2*n):
     z1[i] = op(z1[i-1],z1[i-2]);
     p1[i] = dp*i
    
for i in range(2,300):
     z2[i] = op(z2[i-1],z2[i-2]);
     p2[i] = dp*i
```
