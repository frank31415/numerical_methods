```python
import numpy as np
import math

dt = 0.01
m=1
n = 880
```

```python
r = np.zeros(n)
t = np.zeros(n)
r[0] = 2.01
r[1] = 2.01
#
r1 = np.zeros(n)
t1 = np.zeros(n)
r1[0] = 2
r1[1] = 2
#
r2 = np.zeros(n)
t2 = np.zeros(n)
r2[0] = 1.99
r2[1] = 1.99
```

```python
def nextrp(r1,r0):
    if r1 > 0:
        x = (1-2*m/r1)*2*dt+r0
    else:
        x = 0
    return max(x,0)

def nextrm(r1,r0):
    x = -(1-2*m/r1)*2*dt+r0
    return x
```

```python
for i in range(2,n):
    r[i] = nextrp(r[i-1],r[i-2])
    t[i] = dt*i

for i in range(2,n):
    r1[i] = nextrp(r1[i-1],r1[i-2])
    t1[i] = dt*i

for i in range(2,n):
    r2[i] = nextrp(r2[i-1],r2[i-2])
    t2[i] = dt*i
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 8), dpi=80)
plt.plot(t,r)
plt.plot(t1,r1)
plt.plot(t2,r2)
plt.ylabel('r')
plt.xlabel('t')
plt.show()
```
