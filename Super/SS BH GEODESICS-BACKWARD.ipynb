```python
import math
import numpy as np
```

```python
h = 0.01
m = 1
n = 20
```

```python
def evot(t1,t0,r1):
    x = -2*m/(r1*(r1-2*m))*(1-2*m/r1)*h**2+2*t1-t0
    return x
```

```python
def evor(t2,t0,r1,r0):
    x = 2*r1-r0 - m/r1**3*(r1-2*m)*(t2-t0)**2/4+m*4*h**4*(1-2*m/r1)**2/(t2-t0)**2/(r1*(r1-2*m))
    if x > 0:
        return x
    else:
        return 0
```

```python
r = np.zeros(n)
t = np.zeros(n)
r[0] = 180
r[1] = 180
t[0] = 10
t[1] = 0
```

```python
for i in range(2,n):
    t[i] = evot(t[i-1],t[i-2],r[i-1])
    r[i] = evor(t[i],t[i-2],r[i-1],r[i-2])
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 8), dpi=80)
plt.plot(r,t)
plt.ylabel('t')
plt.xlabel('r')
plt.show()
```
